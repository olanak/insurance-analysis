```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")
```

```python
# Load the dataset
df = pd.read_csv('../data/raw/MachineLearningRating_v3.txt', delimiter='|')

# Display the first 5 rows
df.head()
```

```python
# Check data types and non-null counts
df.info()

# Summarize numerical columns
df.describe()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
# Fill missing values with 'Unknown' for categorical and 0 for numerical
df.fillna({'Citizenship': 'Unknown', 'TotalPremium': 0, 'TotalClaims': 0}, inplace=True)
```

```python
# Convert to datetime
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])

# Extract year and month for time series analysis
df['Year'] = df['TransactionMonth'].dt.year
df['Month'] = df['TransactionMonth'].dt.month
```

```python
# Plot distribution of TotalPremium
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPremium'], bins=30, kde=True)
plt.title('Distribution of Total Premium')
plt.xlabel('Total Premium')
plt.show()
```

```python
# Select numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

# Plot correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Province', y='TotalPremium', data=df)
plt.title('Total Premium Distribution by Province')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by year and month to calculate total premiums
monthly_trend = df.groupby(['Year', 'Month'])['TotalPremium'].sum().reset_index()

# Create a new 'Date' column for plotting
monthly_trend['Date'] = pd.to_datetime(monthly_trend[['Year', 'Month']].assign(day=1))

# Plot the trend
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='TotalPremium', data=monthly_trend)
plt.title('Monthly Total Premium Over Time')
plt.xlabel('Time')
plt.ylabel('Total Premium')
plt.xticks(rotation=45)
plt.show()
```
